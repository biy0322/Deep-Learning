```python
import torch
import torch.nn as nn
```

# Filter 선언 방법 -> Conv2d

- **Conv2d(input channel, output channel, kernel_size, stride, padding)**

- input_ch = 1, output_ch = 1, kernel_size=3
<=> input channel1, ouptut channel도 1, kernel size는 3*3
- 만약, kernel_size = (3,1)이라면, 3*1 kernel

```python
conv1 = nn.Conv2d(1, 12, 2) # 정사각형 모형의 kernel size
print(conv1)

conv2 = nn.Conv2d(1, 12, (2,2)) # 위와 같은 결과
print(conv2)

conv3 = nn.Conv2d(1, 12, (2, 3)) # 직사각형 모형의 kernel size
print(conv3)

# input channel = 3, output channel = 5, kernel size = 11, strid = 4, padding = 1
conv4 = nn.Conv2d(3, 5, 11, stride=4, padding=1)
print(conv4)
```

# 실제로 Data에 Conv Layer에 넣어보고 Output 관찰

- kernel size, stride, padding을 이용해서 output data의 shape 계산

$$shape = \displaystyle \frac{(Input shape) - (kernel size)+ 2*(padding)}{stride} + 1$$

Ex)
1. input data shape = (57,57)
2. Kernel size = (3,3)
3. stride = 2
4. padding = 0

$$shape = \displaystyle \frac{(57,57) - (3)+ 2*(0)}{2} + 1= \displaystyle \frac{(54,54)}{2} + 1= \displaystyle {(27,27)} + 1= \displaystyle {(28,28)}$$

## Example1

- H = 57, W = 57 input 선언(data 갯수 1, channel 3이라 가정)
- Output Channel은 32
- Convolution Layer는 아까와 마찬가지로 kernel size=3, stride=2, padding=0

### data 선언

```python
input_size = torch.Tensor(1, 3, 57, 57) # batch_size = 1, input channel = 3, 
input_size.shape
```

- input channel = 1인 경우, 흑백
- input channel = 3인 경우, 컬러(r, g, b)

### Convolutional Layder 선언

```python
conv = nn.Conv2d(3, 23, 3, stride=2, padding=0)
conv
```

```python
out = conv(input_size)
out.shape
```

## Example3

- input shape = (224, 224)
- Conv Layer : kernel_size = 7*7, stride=4, padding=0
- input channel = 64
- output channel = 128

```python
input_size3 = torch.Tensor(1, 64, 224, 224)
conv3 = nn.Conv2d(64, 128, 7, stride=4, padding=0)

out3 = conv3(input_size3)

print("output data의 shape는 ", out3.shape) ## 55.25 -> 내림을 통해 55
```
